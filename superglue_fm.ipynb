```python
from mlfactory.applications.superglue_inference import match_pair
```

```python
m = match_pair.matcher()
```

Upload images you want to match, here I uploaded 'car.jpg' two times from my local machine and renamed one of the image to 'car_copy.jpg'. You can upload your own image and supply names as shown below. The match() function will return arrays of matching 2D points and will matplotlib enabled, also visualize the matches

Run and visualize the inference !

```python
m.viz_plt = True
import numpy as np
import cv2

im1 = np.array(cv2.imread('car.jpg',0))
im2 = np.array(cv2.imread('car_copy.jpg',0))
m.match(im1,im2)
```
