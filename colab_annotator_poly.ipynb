```python
from mlfactory.datacreators.utils import colab_poly_anot as poly_annotator
```

When initializing the annotator you need to supply a list of all the file names (png or jpg) on which you are annotating, also supply the name of the workspace where the results would be stores as Images and Masks.

When each image appears click on the images consequtively to trace a polygon and when its almost enclosing the object, press 'x', it will use opencv hull approximation to complete the convex polygon annotation. Then it will load the next image and keep doing it until all the images in the list are annotated.

```python
#supply list of image file names , click continuous to make a polygon after its almost enclosing, press x and it will load the next image name in the list
#result would be saved as hull.png
pa = poly_annotator.poly_annotator(["car.jpg"], desired_size = (-1,-1), normalize_bbox = False, workspace = "trial")
pa.run()
```

Since workspace="trial" was supplied, a folder named "trial" would be created in your colab folder space. This folder would contain Images and Masks subfolders and inside them you can double click and see the Image and polygon masks. You can download them and use them for segmentation projects
